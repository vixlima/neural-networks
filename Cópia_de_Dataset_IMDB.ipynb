```python
import urllib.request
import tarfile
import os
```

```python
url = "https://ai.stanford.edu/~amaas/data/sentiment/aclImdb_v1.tar.gz"
file_name = "aclImdb_v1.tar.gz"
```

```python
if not os.path.exists("aclImdb"):
    print("Baixando o arquivo do IMDB (isso pode demorar um pouquinho, são uns 80MB)...")
    urllib.request.urlretrieve(url, file_name)

    print("Descompactando a pasta...")
    with tarfile.open(file_name, "r:gz") as tar:
        tar.extractall()
    print("Pronto! Arquivos extraídos.")
else:
    print("A pasta 'aclImdb' já existe. Não precisa baixar de novo.")
```

```python
import glob

print("Lendo os arquivos de texto...")

caminhos_positivos = glob.glob("aclImdb/test/pos/*.txt")[:100]
caminhos_negativos = glob.glob("aclImdb/test/neg/*.txt")[:100]

textos = []
gabarito = []
```

```python
# Lendo os positivos
for caminho in caminhos_positivos:
    with open(caminho, "r", encoding="utf-8") as arquivo:
        # Pego só os primeiros 1500 caracteres para o modelo não reclamar que o texto é muito grande
        textos.append(arquivo.read()[:1500])
        gabarito.append("POSITIVE")
```

```python
# Lendo os negativos
for caminho in caminhos_negativos:
    with open(caminho, "r", encoding="utf-8") as arquivo:
        textos.append(arquivo.read()[:1500])
        gabarito.append("NEGATIVE")
```

```python
print(f"Pegamos {len(textos)} avaliações de filmes no total para fazer o teste!")
```

```python
from transformers import pipeline
```

```python
print("Baixando o 'cérebro' do Hugging Face (modelo pré-treinado)...")
# Quando a gente não diz qual modelo quer, ele pega um padrão muito bom automaticamente
analisador = pipeline("sentiment-analysis", truncation=True)
```

```python
print("\nAnalisando os 200 textos (pode levar alguns segundos)...")
# A IA lê todos os textos de uma vez aqui!
resultados = analisador(textos)
```

```python
# Vamos contar quantos ele acertou
acertos = 0
```

```python
for i in range(len(textos)):
    # O modelo responde um dicionário tipo: {'label': 'POSITIVE', 'score': 0.99}
    chute_do_modelo = resultados[i]['label']

    if chute_do_modelo == gabarito[i]:
        acertos += 1
```

```python
print("-" * 40)
print(f"O modelo acertou {acertos} de {len(textos)} textos!")
print(f"Acurácia final: {(acertos/len(textos))*100}%")
print("-" * 40)
```

```python
# Mostrando um exemplo na prática pra ver como sai:
print("\nVeja um exemplo prático do primeiro texto:")
print("TEXTO ORIGINAL:", textos[0][:150], "...")
print("O QUE ERA DE VERDADE:", gabarito[0])
print("O QUE O MODELO RESPONDEU:", resultados[0])
```
